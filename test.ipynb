```python
import sqlite3

conn = sqlite3.connect("cities500.db")
c = conn.cursor()

c.execute("""
CREATE TABLE IF NOT EXISTS cities500 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    ascii_name TEXT,
    alt_names TEXT,
    lat REAL,
    lon REAL,
    cc TEXT,
    pop INTEGER
);""")

cities = []

with open("cities500/cities500.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    
    n = len(lines)//100
    i = 1
    
    for l in lines:
        values = l.split('\t')
            
        # values name / ascii name / alt names / lat / lon / cc / pop
        
        name = values[1]
        ascii_name = values[2]
        alt_names = values[3]
        lat = values[4]
        lon = values[5]
        cc = values[8]
        pop = values[14]
        
        cities.append((name, ascii_name, alt_names, lat, lon, cc, pop))

c.executemany("INSERT INTO cities500 (name, ascii_name, alt_names, lat, lon, cc, pop) VALUES (?, ?, ?, ?, ?, ?, ?)", cities)
conn.commit()

conn.close()
```

```python
import pandas as pd

conn = sqlite3.connect("cities500.db")
c = conn.cursor()

c.execute("SELECT * FROM cities500")
results_df = pd.DataFrame(c.fetchall())

print(results_df)

c.close()
```

```python
import sqlite3

conn = sqlite3.connect("users.db")
c = conn.cursor()

c.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    username TEXT,
    email TEXT,
    password TEXT,
    points INTEGER
);""")

c.close()
conn.close()
```
